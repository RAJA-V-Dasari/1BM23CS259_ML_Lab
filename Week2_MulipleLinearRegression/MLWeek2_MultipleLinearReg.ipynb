```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load Diabetes dataset
data = load_diabetes()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

# Use only ONE feature: bmi
X_simple = X[['bmi']]

# Split dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(
    X_simple, y, test_size=0.2, random_state=42
)

# Create and train Linear Regression model
model_simple = LinearRegression()
model_simple.fit(X_train, y_train)

# Predict
y_pred = model_simple.predict(X_test)

# Evaluate
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("Simple Linear Regression (Diabetes dataset)")
print("RMSE:", rmse)
print("Intercept:", model_simple.intercept_)
print("Slope:", model_simple.coef_[0])
```

```python
# Use all features for Multiple Linear Regression
X_multi = X

# Split dataset into train and test
X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(
    X_multi, y, test_size=0.2, random_state=42
)

# Create and train Linear Regression model
model_multi = LinearRegression()
model_multi.fit(X_train_m, y_train_m)

# Predict
y_pred_m = model_multi.predict(X_test_m)

# Evaluate
rmse_m = np.sqrt(mean_squared_error(y_test_m, y_pred_m))

print("\nMultiple Linear Regression (Diabetes dataset)")
print("RMSE:", rmse_m)
print("Intercept:", model_multi.intercept_)
print("Coefficients:", model_multi.coef_)
```
