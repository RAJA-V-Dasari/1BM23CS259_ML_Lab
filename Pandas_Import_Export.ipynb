```python
import pandas as pd

data = {
    "ID": range(1, 101),
    "Name": [f"Student_{i}" for i in range(1, 101)],
    "Age": [18 + (i % 5) for i in range(100)],
    "Marks": [50 + (i % 51) for i in range(100)]
}

df = pd.DataFrame(data)

df.to_csv("students_100.csv", index=False)
print("100 rows exported successfully!")

imported_df = pd.read_csv("students_100.csv")

print("\nImported Data:")
print(imported_df.head(10))
```
